```python
dis = []
models = []
max_k = 30
for k in range(1,max_k+1):
    print(k)
    model = KMeans(k, max_iter=10)
    model.fit(img)
    dis.append(model.inertia_)
    models.append(model)
```

```python
K = range(1,len(dis)+1)
plt.plot(K, dis)
```

```python
def process_img(models,i):
    model = models[i]
    processed_img = np.zeros_like(img)
    for i in range(model.cluster_centers_.shape[0]):
        processed_img[model.labels_ == i,:] = model.cluster_centers_[i]
    processed_img = processed_img.astype("int")
    processed_img = processed_img.reshape((w,h,3))
    return processed_img
```

```python
processed_img = process_img(models,25)
large_processed_img = cv.resize(processed_img, (750,750), interpolation = cv.INTER_NEAREST)
plt.imshow(processed_img)
```
